```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('../input/train.csv')
test = pd.read_csv('../input/test.csv')
full_data = pd.concat([train, test], ignore_index=True,sort=False)
```

```python
# Sort columns by number of 0
count_col_0 = full_data.apply(lambda x: np.where(x==0)[0].shape[0],axis=0)
indices_col_0 = np.argsort(count_col_0)
data_sorted_col_0 = full_data.iloc[:,indices_col_0]
```

```python
# Sort rows by number of 0
count_row_0 = full_data.apply(lambda x: np.where(x==0)[0].shape[0],axis=1)
indices_row_0 = np.argsort(count_row_0)
data_sorted_row_0 = data_sorted_col_0.iloc[indices_row_0,:]
```
