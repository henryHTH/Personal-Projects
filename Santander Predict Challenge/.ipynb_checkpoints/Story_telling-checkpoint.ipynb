```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from RF_feature_selection import *
import warnings
warnings.filterwarnings('ignore')
```

```python
df,target,df_info = EDA('train.csv')
```

```python
data = pd.read_csv('train.csv')
```

```python
plt.figure(figsize=(30, 10))
plt.hist(target,bins=500, alpha = 0.5, histtype='bar') # Transactions tend to in X0000 (int)
plt.xlabel('Transaction')
plt.ylabel('Count')
plt.title('Target Distribution')
plt.show()
```

```python
corrs = df_info.loc['corr',df.columns]
plt.hist(corrs,bins = 50,alpha = 0.5, histtype='bar')
plt.suptitle('Feature vs Target Correlation')
plt.ylabel('Feature Count')
plt.xlabel('Correlation Level')
plt.show()
```

```python
#check feature correlation with target
df_info = df_info.sort_values(by='corr', ascending=False, axis=1)
high_corr_col = df_info.columns[np.where(df_info.loc['corr']>=0.2)]
high_corr_col = [x for x in df.columns if x in high_corr_col]
df_high_corr = df[high_corr_col]
df_info[high_corr_col].sort_values(by='zero_rate', ascending=False, axis=1)
```

```python
from string import ascii_letters
top_cor_feature = df_info[high_corr_col].sort_values(by='zero_rate', ascending=False, axis=1).columns[:25]
sns.set(style="white")
# Compute the correlation matrix
corr_top = df[top_cor_feature].corr()
#corr_high = df[high_corr_col].corr()
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_top, dtype=np.bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(240, 10, n=9,as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_top, mask=mask, cmap=cmap, vmax=1, center=0.5,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
#Check Top 5 feature correlations with target
g = sns.pairplot(data[['target'] + list(top_cor_feature[:5])])
```
