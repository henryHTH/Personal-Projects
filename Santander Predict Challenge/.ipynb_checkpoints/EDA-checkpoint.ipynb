```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble.partial_dependence import plot_partial_dependence
from sklearn_gbmi import *
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
RANDOM_SEED = 137
TRAIN_FRACTION = 0.9
```

```python
#train boosting tree for interation detection
X_train, X_test, y_train, y_test = train_test_split(df, data.target, test_size=0.2, random_state=RANDOM_SEED)
gbr_1 = GradientBoostingRegressor(random_state = RANDOM_SEED)
gbr_1.fit(X_train,y_train)
scores1 = sorted(gbr_1.feature_importances_,reverse=True)
scores1 = {df.columns[i]:scores1[i] for i in range(len(scores1)) if scores1[i]!=0}
scores1 = {k:v for k,v in sorted(scores1.items(), key=lambda x:x[1],reverse=True)}
y_pred1 = gbr_1.predict(X_test)
print ("R-squared for Train: %.2f" %gbr_1.score(X_train, y_train))
print ("R-squared for Test: %.2f" %gbr_1.score(X_test, y_test)) 
print ("Mean_squared_error for Test: %.2f" %mean_squared_error(y_pred1, y_test)) 
print ("Mean_absolute_error for Test: %.2f" %mean_absolute_error(y_pred1, y_test)) 
```

```python
high_corr_col = df_high_corr.columns
match_importance = {}
combine_col = [key for key in scores1.keys()].copy()
for col in high_corr_col:
    if col in scores1:
        match_importance[col] = (data[['target',col]].corr().values[0][1],scores1[col])
#match_importance
    else:
        combine_col.append(col)
#combine_col
```

```python
from sklearn.ensemble import RandomForestRegressor 
from sklearn.feature_selection import SelectFromModel
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, data.target, test_size=0.2, random_state=RANDOM_SEED)
rf = RandomForestRegressor(n_estimators = 20,random_state=RANDOM_SEED)
sel = SelectFromModel(rf)
rf.fit(X_train, y_train)
sel.fit(X_train, y_train)
selected_feat= X_train.columns[(sel.get_support())]
#len(selected_feat)
print(selected_feat)
```

```python
plt.figure(figsize=(30, 10))
column_loc = [df.columns.get_loc(x) for x in selected_feat]
plt.bar(selected_feat,rf.feature_importances_[column_loc],align='center', alpha=0.5)
plt.xticks(rotation=90)
plt.show()
```
